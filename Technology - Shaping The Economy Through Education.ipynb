### Background Information
Since the advent of information technology, mediums such as the internet, tv and film have changed the way we live. New ways of sharing ideas have led to a mass distribution of information. This ability to effectively share information and thus market have a positive effect on the economy. 

Technology such as the internet and computers provide instantaneous access to the rest of the world allowing us to market, communicate, facilitate ideas, create and hence create economic products through this ever-globalizing world economy. At your fingertip lies data to allow us to analyze profitability, the stock market and more. 

Modern day developments have been made possible through the technology field. Historic business models in long-standing industries have been disrupted due to growing technological fields and their economic dominance. Some examples include the services industries due to emerging companies such as Uber and Lyft and conglomerate like Amazon. A new era is upon us and certain skills are necessary to thrive in this environment. Investment in the education of these burgeoning fields will develop individuals with the skills to innovate and benefit the economy. The ability to innovate has helped the economy grow hence a new to help students develop skills in these industries. 

Information technology has a lot to offer us and with new technological innovations on the horizon, the reach of the internet will only continue to grow, providing an opportunity for the government to welcome the growing platform. According to Manyika and Roxbourgh of McKinsey Global Institute, the "internet accounts for on, average, 3.4 percent of GDP across the large economies that make up 70 percent of global GDP" [1]. 

Technology companies and employees are desirable to state policy makers. They generally are high earning and consume fewer resources than other industries. Technology companies also scale quickly and can add considerable positive impact to a state’s budget in a few short years. As a result, states do a lot to attract technology companies and employees; tax incentives, infrastructure investments, and educational investments.  

However, these methods are not all equal. Tax breaks can be controversial and hard to sell to constituents with sometimes unclear benefits. Infrastructure is often a bar to clear rather than scalable solution to increasing economic activity in the technology sector.  

One area where states have direct control is over the investment in technology (computer science, etc.) departments at public universities. These universities not only graduate students with the required skills to work in tech or start a company but often the university serves to create an entire ecosystem for the technology sector to thrive.

### Problem Statement
States have long had a desire to attract employees in the field of technology. They use a series of incentives and investments but it’s not always clear how effective each tool is.  

One tool directly falls under the power of the state – investment in public higher education institutions. However, policy makers need to understand if a direct link exists between investment and future economic activity within their state, and how much of a return can be expected. Hence a need to answer the question; whether, and to what extent, does increasing state-level investment in technology departments of higher education institutions lead to increased economic activity in the technology sector for a state? Is this a risk policy makers are interested in taking considering the current information technology landscape. Have past investments helped shaped this lanscape.  

Answering these questions can give policy makers a roadmap on where and how much to invest in universities for their computer science and other related technology departments. Our supporting questions get to the details of how a policy might be formed. It’s not just the volume of students but quality as well. Similarly, the advanced degree holders potentially hold more value than undergraduates.  

Additionally, this could provide concrete evidence of the benefits and return on invested taxpayer dollars which can be used to sell constituents and other policy makers on increased investment.

### Hypothesis
We expect to find a significant though small correlation between investment and economic activity. Likely, there will be a low R-squared value since there are many factors which play into the level of economic activity in an individual state’s technology sector.  

Accounting for the quality of students and mix of graduate students will add some small amount of explanatory power but it will be difficult to draw out causality which likely goes both ways. E.g. States with more investment attract better students and better students attract more investment.

#### Data Source 3: FIPS Data
The dataset from the Bureau of Economic Analysis (BEA) helps to map each institution to its state. Within the College Scorecard data source is a FIPS code. This FIPS dataset helps us identify the state an educational institutions resides within. 



#### Data Source 4: State Census Population 1990-2020
The dataset from the US Department of Agriculture (USDA) contains 10 year census data for all 50 states plus Washington DC, US total, and Puerto Rico from 1990 to 2020. This dataset allows us to normalize our state GDP dataset and the number of students in educational institutions by state and year. 



#### Data Source 5: IPEDS
The Inetgrated Post Seconadary Education System (IPEDS) dataset from the National Center for Education Statistics contains data for degrees given by major by year, going back 20 years. The dataset includes undergraduate and graduate degrees. The degrees of interest to us were Computer Science and Information Sciences. This dataset will be joined with the College Scorecard dataset and linked by year and state to be aggregated over each respective section. 



#### Data Cleaning Process
Initial intentions with this project were to perform dimensionality reduction on the datasets but considering the vast amount of data unrelated to the scope of the project, individual rows were selected based on their relevance and importance to the project by observing FIPS codes, SIPS codes, and various states. 

We joined the annual GDP By State By Industry, College Score Card and FIPS datasets on common variable and manually filtered down variables based on applicability to overall project object to create a merged dataframe. Within the joined dataset, we aggregated student degree totals by state and year. Next we filtered the dataset to relevant years. We would then join the census population data and normalize GDP and number of technology students.

#### Regression Analysis
Majority of the modeling done was linear regression to identify whether a relationship exists between the number of students in technology majors and a state's technology GDP output. In this regression section, we will include the models, as well as the visualizations for modeling assumptions (normality and independence).

##### Pre-First Model

For our first model, we ran a linear regression model with GDP per capita as the response variable and number of students per capita as the predictor variable. The adjusted R-squared value is 0.6287 which spells good news but when calculating Cooks Distance to validate our assumptions, we see that there is a leverage point. Washington DC is an outlier. 



##### First Model: Base Regression
Since Washington DC is an outlier, we will remove that data point and create a new regression model with the new dataframe. 


This model performs decently. The model assumptions hold true. The coefficients are significant but they R-squared value is 0.2187. This indicates that only a small amount of the variation in technology GDP per capita can be explained by the predictor variable of number of technology students per capita. Considering our hypothesis that we expect a low R-squared value, this result seems fair since there are likely to be a lot factors within a states technology GDP.

### Conclusions

#### Challenges

Some challenges we faced during this project is that GDP is influenced by many factors. The technology industry is also still quite young so our data was limited 20 to 30 years of data. Within 10 years, it would be great to have additional data of how investment can impact the future.

#### Further Exploration
Some ideas of interest for the future include analyzing the percent change in GDP as a response to the percent change in technology students year over year. It would also be interesting to compare other industry sectors to understand how each may affect GDP. It would also be interesting to test the quality of students graduating against GDP growth. 

#### Key Takeaway

One thing for all of our regression models is the correlation doesn't equate to causation. As stated in the hypothesis, many factors play a part in the technology GDP of a state. While we may see a strong correlation between technology GDP and number of technology students, other factors could be at play. 

States that might be more willing to invest in technology programs and have a strong university system may have other technology infrastructures contributing to high technology GDP. The number of technology students is more likely correlated with other factors that all contribute to the infrastructure of technological economic growth.

Despite this, our results are still meaningful. Private non-profit schools and private for-profit schools were quite significant and high coefficients. Likewise, based on their high coefficients and significance, Masters and PhD programs might be worthy investments. Again as stated earlier based on our low R-squared values, number of students in technology programs isn't likely to explain the variation in technology GDP per capita on its own. Other factors may be at play. 

### Article Review
In Hanushek and Woessmann's Education and Economic Growth, the importance of education in promoting economic growth is discussed. While technology isn't specifically addressed in the article, one of the reasons education is viewed as an indication of economic well-being is because it can "increase the innovative capacity of the economy, and the new knowledge on new technologies, products, and processes promotes growth"[2]. The value of education lies in the ability to apply knowledge to our lives and the world around us. Research indicates that education is important in promoting development with an emphasis on higher education for encouraging innovation. 

Hanushek and Woessman also address historical methods of correlating the economic impact of education. These methods included estimating "cross-country growth regressions where average annual growth in gross domestic product(GDP) per capita over several decades is expressed as a function of measures of schooling and a set of other variables deemed important for economic growth"[2]. These methods suggested that there was a positive correlation between numbers of years of schooling and economic growth. These historic methods and subsequent studies though often led to controversy on whether the quality of schooling as opposed to just number of years played a larger part in economic growth. For this project, we are looking primarily at the economic impact of higher education in technology. Our primary focus on higher education in a particular industry and not necessarily number of years of schooling can help us address the issue of quality of schooling. 

In the current information technology era, we've seen many industries adopt technological changes in an effort to help them innovate, develop, grow economically. As industries make the shift towards incorporating information technology, the emerging workforce needs to be prepared with the skillset to allow them to contribute to this growing industry. With the advent of NASA and space flights came aerospace engineering to develop the future that will reach our next frontier and a growing industry. Similarily, with developments in information technology and its economic impacts, investments in technology programs to educate the future should be considered.
